## Enriquecer el set de datos

Vamos a aumentar los datos presentes en el conjunto dado con un dataset similar: las publicaciones de la plataforma AirBnB en Melbourne en el ao 2018. El objetivo es estimar con mayor presicin el valor del vecidanrio de cada propiedad.

Fuente: (https://www.kaggle.com/tylerx/melbourne-airbnb-open-data?select=cleansed_listings_dec18.csv) de *scrapings* del sitio realizado por [Tyler Xie](https://www.kaggle.com/tylerx), tambin disponible en una competencia de Kaggle.

```python
# data source:
# https://www.kaggle.com/tylerx/melbourne-airbnb-open-data?select=cleansed_listings_dec18.csv
interesting_cols = [
  'description', 'neighborhood_overview',
  'street', 'neighborhood', 'city', 'suburb', 'state', 'zipcode',
  'price', 'weekly_price', 'monthly_price',
  'latitude', 'longitude', ]
airbnb_df = pd.read_csv('https://cs.famaf.unc.edu.ar/~mteruel/datasets/diplodatos/cleansed_listings_dec18.csv', usecols=interesting_cols)
```

Para poder realizar esta combinacin de datos, tenemos que encontrar una columna que contenga informacin en comn. Utilizaremos el zipcode.

* Qu desventajas tiene utilizar el zipcode para agregar informacin sobre la vecindad de cada propiedad en venta? Se corresponde con lo que entendemos intuitivamente como vecindad?
* Qu otros mtodos podran utilizarse?

```python
airbnb_df.zipcode.value_counts()[:5]
```

 Sin embargo, este zipcode tiene tipos diferentes, que tendremos que estandarizar antes de continuar con la combinacin de los datos.

```python
airbnb_df['zipcode'] = pd.to_numeric(airbnb_df.zipcode, errors='coerce')
```

### Valores en comn

Antes de unir los conjuntos de datos, tenemos que asegurarnos de que tengamos suficientes registros en comn como para agregar informacin relevantes.

```python
intersection = numpy.intersect1d(airbnb_df.zipcode.values, mel_df.Postcode.values, assume_unique=False)
print("Airbnb unique zipcodes", len(airbnb_df.zipcode.unique()))
print("Sales unique zipcodes", len(mel_df.Postcode.unique()))
print("Common zipcodes", len(intersection))
```

```python
print('Registros en mel_df con el cdigo postal correspondiente con Airbnb df',
      mel_df.Postcode.isin(intersection).sum() / len(mel_df))
print('Registros en Airbnb_df con el cdigo postal correspondiente en mel_df',
      airbnb_df.zipcode.isin(intersection).sum() / len(airbnb_df))
```

## Operacin merge

Es hora de combinar ambos conjuntos de datos. Una opcin frecuentemente tentadora es aplicar la operacin merge sin analizar demasiado la situacin. Sin embargo, esta operacin puede ocasionar una **explosin exponencial** de nuestro conjunto de datos!

Por qu la siguiente operacin genera un DataFrame con 2 millones de filas? 

```python
## Incorrecto!
len(mel_df.merge(airbnb_df, how='left',
                  left_on='Postcode', right_on='zipcode'))
```

Recordemos que nuestro objetivo es agregar informacin sobre el valor del vecindario de la propiedad. Qu informacin del conjunto de airbnb nos sirve para representar esto?

* Promedio de `price` por zipcode
* Promedio de `weekly_price` por zipcode
* Promedio de `monthly_price` por zipcode

Queda a interpretacin de ustedes decidir cul de todos ellos utilizar.

```python
relevant_cols = ['price', 'weekly_price', 'monthly_price']
```

Cuando queremos agregar un nico dato luego de agrupar un DataFrame, la operacin es bastante sencilla. Slo tienen que tener en cuenta que el nombre de la columnas ser el mismo, aunque su valor ahora est agregado.

```python
# Option 1: no tan eficiente
# La agregacin se calcula en todas las columnas
# Filtramos luego slo las columnas que queremos.
# airbnb_df.groupby('zipcode').mean()[relevant_cols].reset_index()[:5]

# Opcin 2: ms eficiente
#Primero filtramos las columnas que queremos y luego solo agregamos
# Tenga cuidado de incluir tambin la columna de agrupacin
airbnb_df[relevant_cols + ['zipcode']]\
  .groupby('zipcode').mean().reset_index()[:5]
```

Agrupar y agregar utilizando ms de un tipo de operacin es un poco ms complicado. Pandas nos provee de muchas formas de escribir la misma operacin. Ninguna de ellas es muy intuitiva, y usualmente requiere renombrar las columnas para darles nombres ms significativos.

```python
relevant_cols = ['price', 'weekly_price', 'monthly_price', 'zipcode']
## Option 1
# Pass a dictionary where the keys are the original columns to aggregate and
# the values are the operations (or list of operations).
airbnb_price_by_zipcode = airbnb_df[relevant_cols].groupby('zipcode')\
  .agg({'price': ['mean', 'count'], 'weekly_price': 'mean',
        'monthly_price': 'mean'})\
  .reset_index()
# Flatten the two level columns
airbnb_price_by_zipcode.columns = [
  ' '.join(col).strip()
  for col in airbnb_price_by_zipcode.columns.values]
# Rename columns
airbnb_price_by_zipcode = airbnb_price_by_zipcode.rename(
    columns={'price mean': 'airbnb_price_mean',
             'price count': 'airbnb_record_count',
             'weekly_price mean': 'airbnb_weekly_price_mean',
             'monthly_price mean': 'airbnb_monthly_price_mean'}
)

## Option 2
# Pass as argument name the new name of the column, and as value a tuple where
# the first value is the original column and the second value is the operation.
# airbnb_price_by_zipcode = airbnb_df[relevant_cols].groupby('zipcode')\
#   .agg(airbnb_record_count=('price', 'count'),
#        airbnb_price_mean=('price', 'mean'),
#        airbnb_weekly_price_mean=('weekly_price', 'mean'),
#        airbnb_monthly_price_mean=('monthly_price', 'mean'))\
#   .reset_index()

airbnb_price_by_zipcode[:3]
```

```python
merged_sales_df = mel_df.merge(
    airbnb_price_by_zipcode, how='left',
    left_on='Postcode', right_on='zipcode'
)
merged_sales_df.sample(5)
```
