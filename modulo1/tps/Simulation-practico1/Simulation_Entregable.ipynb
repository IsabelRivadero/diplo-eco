```python
# PARA LA ELABORACION DEL DATASET, EL CAMPO "OCUPACION" ES DE CARACTER ABIERTO. NECESITAMOS REAGRUPAR TODAS LAS OCUPACIONES INGRESADAS
# DEFINIMOS UNA FUNCION PARA EL MATCHEO DE LOS DATOS CON UNA LISTA DE REFERENCIA.
new=[]
proporcion=[]

def matcheo(mal,bien):
    for i in mal:
        x=process.extractOne(i,bien)
        new.append(x[0])
        proporcion.append(x[1])
    return new, proporcion

mal=simu["Ocupación"].dropna().values
bien=['Medico','Enfermero','Paramedico', 'Bombero/Rescatista/Militar','TEM','Otro']

corregido, rango = matcheo(mal,bien)
simu["Ocupación_match"]= pd.Series(corregido)

simu["Acercamiento"]= pd.Series(rango) # Se generan dos nuevas columnas una para la ocupación corregida y otra para el porcentaje de macheo
simu.sample(5) # Mostramos una pequeña porcion del set para ver como se comporta y si existen incoherencias
```
