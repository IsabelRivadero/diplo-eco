# Histogramas para variables numéricas

```python
fig, axes = plt.subplots(nrows=2, figsize=(16, 10))
sbn.histplot(df1["Ing"], bins=6, ax=axes[0], color='gray')
axes[0].axvline(df1["Ing"].mean(), color='red',
            linestyle='--', label='Media')
axes[0].axvline(df1["Ing"].median(), color='blue',
            linestyle='-.', label='Mediana')

sbn.histplot(df1["Aportes"], bins=6, ax=axes[1], color='gray')
axes[1].axvline(df1["Aportes"].mean(), color='red',
            linestyle='--', label='Media')
axes[1].axvline(df1["Aportes"].median(), color='blue',
            linestyle='-.', label='Mediana')

axes[0].legend()
sbn.despine()
```

* La **mediana** se calcula:
  1. Ordenar las realizaciones tal que $x_j \leq x_{j+1}$ 
  2. Si la cantidad de datos $n$ es impar, la mediana se corresponde con el valor que está a la mitad de la serie ordenada: $Me = x_{\lfloor n / 2 \rfloor +1}$
  3. Si la cantidad de datos $n$ es par, la mediana es el promedio de los dos valores centrales: $Me = \frac{1}{2} (x_{ n/ 2 } + x_{ (n / 2) +1})$

###Medidas de tendencia no central 

*   Un **Fractil** es el valor de la variable que está en un orden k y tiene una proporción de observaciones $k/s$


donde $k$ es el orden del fractil y $s$ cantidad de divisiones de la distribución

*   Cuartiles (4 partes): $s=4$
*   Deciles (10 partes): $s=10$
*   Percentiles (100 partes): $s=100$


  1. Ordenar las realizaciones tal que $x_j \leq x_{j+1}$ 
  2. Encontrar el orden o punto de posicionamiento (para los cuatiles  $\frac{k}{4}*(n+1)$
  3. Luego se calcula el valor del cuartil
