##Análisis de datos categóricos

###Tablas de frecuencias para datos categóricos


Una tabla que resume el comporamiento de los datos se denomina tabla de distribución de frecuencias y muestra cómo se distribuyen las unidades de observación entre las categorías.


Una tabla de frecuencias para variables categóricas es una tabla que asocia cada categoría de la variable una frecuencia 


*   El número de veces que se repite cada categoría de la variable se denomina frecuencia absoluta.

   



Simbólicamente: 

$0\leq n_i \leq n$

$\sum_i^n n_i=n $

$k\leq n $


*   La frecuencia relativa de una categoría es la proporción de veces que ocurre dicha categoría.


$h_i = \frac{n}{n_i}$

$0\leq h_i \leq 1$

$\sum_i^n h_i=1 $




*   La frecuencia acumulada representa la cantidad o proporción de casos asociados a una categoría y todas categorías anteriores o "menores" (solo tiene signifiado si la veriable está medida en escala ordinal)




Frecuencias absolutas acumuladas

$0\leq N_j \leq n$

$N_j=\sum_{i=1}^j n_i$

$N_k= n $

Ejemplo

$N_1= n_1$ 

$N_2= n_1 + n_2$

$N_3= n_1 + n_2+ n_3$


Frecuencias relativas acumuladas

$0\leq H_j \leq 1$

$H_j=\sum_{i=1}^j h_i$

$H_k= 1 $

Otra forma de calcular medio complicada pero esta relacionada con la funcion anterior

```python
#Dataframe que contendra las clases únicamente
#Sacamos los datos únicos de data
lisS = df1["Satisf"].unique()
#Creamos el dataframe que contendrá las clases
clasSat = pd.DataFrame(lisS, columns=["Satisf"])
#Observamos las clases que tenemos
clasSat
```

```python
#Tabla de frecuencias absolutas
#Obtenar las frecuencias absolutas de cada clase
faSatis= pd.crosstab(index=df1["Satisf"], columns = "fa")
#Creamos una lista con los valores de las frecuencias
li = faSatis.values
#Agregamos la columna al dataframe (clasSat es el data frame que tiene las clases)
clasSat["fa"] = li
#Observamos
clasSat
```

```python
#Tabla de frecuencia relativa
frSatis = 100 * faSatis["fa"] / len(df1['Satisf'])
lis = frSatis.values
#Agregamos nueva columna de frecuentas relativas
clasSat["fr"] = lis
#mostramos el data frame que tiene las clases con el agregado de la frecuencia relativa
clasSat
```

```python
#Tabla de frecuencia acumulada Continuamos con la columna de frecuencias absolutas acumuladas, 
#ésta en cada renglón se va sumando el dato de la columna de frecuencias absolutas, 
#como se observa de la siguiente manera.

#Obtenemos las frecuencias absolutas acumuladas
#Sacamos una lista de los valores donde obtendremos la Fa
Fa = clasSat["fa"].values
Fa
```

```python
#Recorremos la lista para ir creando una nueva lista con las sumas
#Obtener la frecuencia absoluta acumulada
a =[] 
b=0
for c in Fa:
  b=c+b
  a.append(b)

#Agregamos la nueva columna Fa al Dataframe dat
clasSat["Fa"]= a 
clasSat
```

```python
Hi = clasSat["fr"].values
Hi
```

```python
#Recorremos lista para obtener la frecuencia relativa acumulada
Hi = clasSat["fr"].values
Hi
#Obtenemos Hi
e = []
b = 0
for c in Hi:
  b = c + b
  e.append(b)
#Agregamos la nueva columna Hi al Dataframe
clasSat["Hi"] = e
clasSat
```

##Medidas de posición 

###Medidas de tendencia central 


* La **media muestral** (aritmética) o promedio se calcula como:

$ M(x)=\bar{x} = \frac{1}{n} \sum_{i=1}^n x_i $


Propiedades:

$M(c)=c$

$M(c*x)=c*M(x)$

$M(c+x)=c+M(x)$

$M(x+y)=M(x)+M(y)$

$ \sum_i^n (x_i-M(x))=0$

$ \sum_i^n (x_i-M(x))^2 -> mín$

* La **mediana** se calcula:
  1. Ordenar las realizaciones tal que $x_j \leq x_{j+1}$ 
  2. Si la cantidad de datos $n$ es impar, la mediana se corresponde con el valor que está a la mitad de la serie ordenada: $Me = x_{\lfloor n / 2 \rfloor +1}$
  3. Si la cantidad de datos $n$ es par, la mediana es el promedio de los dos valores centrales: $Me = \frac{1}{2} (x_{ n/ 2 } + x_{ (n / 2) +1})$

###Medidas de tendencia no central 

*   Los **Fractiles** es el valor de la variable que está en un orden k y tiene una proporción de observaciones $k/s$

*   Cuartiles (4 partes): $s=4$
*   Deciles (10 partes): $s=10$
*   Percentiles (100 partes): $s=100$

  2. Encontrar el orden o punto de posicionamiento (para los cuatiles  $\frac{k}{4}*(n+1)$
