```python
# Manejo de dataframes
import pandas as pd

# operaciones numéricas
import numpy as np

#Graficos
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import seaborn as sns
import openpyxl
import matplotlib.font_manager
from matplotlib import style
style.use('ggplot') or plt.style.use('ggplot')
```

```python
df = pd.read_excel("Banco.xlsx",  parse_dates=True, engine="openpyxl")

df
```

```python
df=df.drop(["id"],axis=1)
df
```

```python
numericas= df[["ANTIGÜEDAD","SALDO_TOTAL","SALDO_RESUMEN", "CANT_TARJETAS","EDAD","CUENTA_BASICA","INGRESOS_NETOS","MAXIMA_MORA","ENDEUDAMIENTO","SIT_MAYOR1","SEGMENTO","SITUACION","DIRTY" ]]
numericas
```

```python
numericas.describe()
```

```python
#matriz de correlación
corrnumericas = numericas.corr(method='pearson')
corrnumericas
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(12, 12))

sns.heatmap(corrnumericas,
    annot     = True,
    cbar      = True,
    annot_kws = {"size": 8},
    vmin      = -1,
    vmax      = 1,
    center    = 0,
    cmap      = sns.diverging_palette(20, 250, n=200),
    square    = True,
    ax        = ax
)

ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation = 45,
    horizontalalignment = 'right',
)

ax.tick_params(labelsize = 10)
plt.title('Matriz de correlación')
plt.show()
```
