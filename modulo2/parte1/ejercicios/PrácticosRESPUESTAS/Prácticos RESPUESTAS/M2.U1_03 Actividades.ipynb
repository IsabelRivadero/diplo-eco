# PRÁCTICO 12

Realizar un gráfico de barras con las siguientes características

Eje x: Nombre Provincia (corregir Tierra del Fuego)

Eje y: Densidad Poblacional (pob_provncias / sup_provincias)

Label x: Sin Label X

Label y: Densidad Poblacional (Hab/Km2)

Title: Densidad Poblacional de las Provincias Argentinas

Crear un gráfico de barras

```python
# Importar pyplot y numpy
import matplotlib.pyplot as plt
import numpy as np
# Datos
pob_provincias = [15625084, 273964, 509108, 638645, 3308876, 1214441, 1738929, 318951, 874006, 3194537, 367828, 1055259, 585126, 333642, 681055, 930911, 1235994, 432310, 530162, 673307, 1101593, 1448188, 127205]
sup_provincias = [307571, 243943, 224686, 203013, 165321, 155488, 148827, 143440, 136351, 133007, 102602, 99633, 94078, 89680, 89651, 88199, 78781, 76748, 72066, 53219, 29801, 22524, 21571]
nom_provincias = ["Buenos Aires", "Santa Cruz", "Chubut", "Río Negro", "Córdoba", "Salta", "Mendoza", "La Pampa", "Santiago del Estero", "Santa Fe", "Catamarca", "Chaco", "Neuquén", "La Rioja", "San Juan", "Corrientes", "Entre Ríos", "San Luis", "Formosa", "Jujuy", "Misiones", "Tucumán", "Tierra del Fuego, Antártida e Islas del Atlántico Sur"]

nom_provincias[-1] ="Tierra del Fuego"
den_poblacional= np.array(pob_provincias) / np.array(sup_provincias)

hist=plt.bar(nom_provincias, den_poblacional)
plt.xlabel("Sin Label X")
plt.ylabel("Densidad Poblacional (Hab/Km2)")
plt.title("Densidad Poblacional de las Provincias Argentinas")
plt.xticks(rotation = 60, ha="right")
print(hist) #el argumento "ha" es la alineacion del texto con respecto a la división del eje x
```
